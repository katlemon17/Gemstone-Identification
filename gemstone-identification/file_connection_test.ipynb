# File connection test

```python
from google.cloud import storage
```

```python
# set up access to bucket
client = storage.Client(project = 'gemstone-identification-302822')
```

```python
buckets = client.list_buckets()
```

```python
print("Buckets in {}:".format(client.project))
for item in buckets:
    print("\t" + item.name)
```

```python
bucket = client.get_bucket('gemstone-data')
```

```python
blobs = bucket.list_blobs()

print("Blobs in {}:".format(bucket.name))
for item in blobs:
    print("\t" + item.name)
```

```python
blob_name = 'gemstone-data/test/Alexandrite/alexandrite_18.jpg'
blob = bucket.blob(blob_name)
```

```python
blob
```

```python
blob = bucket.get_blob(blob_name)

print("Name: {}".format(blob.id))
print("Size: {} bytes".format(blob.size))
print("Content type: {}".format(blob.content_type))
print("Public URL: {}".format(blob.public_url))
```

```python
output_file_name = "alexandrite_18.jpg"
blob.download_to_filename(output_file_name)
```
